__Exercicio 1__ - Demonstre o impacto da quantidade de pontos para a geração de gráficos no tempo de processamento e na resolução do gráfico gerado. Dica: use o comando %timeit no jupyter.

```python
def x(t): return t**2 # definindo uma função f(t) = t^2
```

__Resposta__: Podemos constatar que para uma mesma quantidade de 100 loops o tempo de processamento a saida dos gráficos é diretamente proporcional a quantidade de pontos. para o gráfico 1 com 10 pontos o tempo foi de 4.17 ms ± 509 µs já o gráfico 2 com 10000 pontos o tempo de processamente foi de 5.54 ms ± 938 µs.

## 2. Geração de sinais discretos  

A função range representa o intervalo entre dois tempos consecutivos com ajuste de uma passo. Já função stem possui parametros de entrada dados pelo eixo x ou seja, o tempo t e o eixo y. O eixo x quando referenciado
são adotados com os mesmos valores do eixo y. 

```python
# Construindo gráfico de sinal discreto

t= np.arange(-2, 2.2, .2) # gerando o vetor que irá representar o espaço de tempo t (inicio, fim, passo)
y = x(t)
plt.stem(y)
```

 A função steam retorna uma tupla com três elementos: markerline , que são os marcadores (as bolinhas), stemlines , que são as linhas ligadas aos marcadores e a baseline , que é a linha de base (que no gráfico aparece vermelha).

### 3.1. Sinal seno e cosseno contínuo

```python
# Implementando sin da biblioteca Numpy

t = np.linspace(-5,5,200) # gerando o vetor que irá representar o espaço de tempo t (inicio, fim, numero de pontos)
y = np.sin(2*t)           # f(x) = sen(2t)
plt.title('sen(2t)')
plt.plot(t, y)            # construindo gráfico
```

__Exercicio 12 -__ Como gerar a função sinal e o sinal retangular a partir de operações com sinais degrau unitários? Implemente e explique sua implementação.


```python
def degrau(x):
    if x < 0:
        return -1 
    elif x > 0:
        return 1
    else:
        return 0

t = np.linspace(-1, 1, 250)
y = [degrau(i) for i in t]
plt.plot(t, y)
```

__Exercicio 17 -__ Por definição, a função rampa cresce indefinidamente para valores acima de 0. Porém, diversas aplicações usam versoes dela com limitação de amplitude. Uma das versões mais comuns é a chamada rampa unitária, ilustrada na figura abaixo. Implemente uma função que gere o gráfico abaixo, e explique sua implementação.


![alt text](image/graficoSinais2.png)

```python
def rampa(t): 
    return np.maximum(0, t)

def sat(t, x1, x2):
    if t <= x1:
        return x1
    elif t >= x2:
        return x2
    else:
        return t
    
base_inf, base_sup = 2, 5
t = np.linspace(0, 10, 250)
y = rampa(t)
y = [sat(i, base_inf, base_sup) for i in y]
plt.plot(t, y)
```

```python
def porta(t,tau): 
    return (np.abs(t) <= tau/2)

t1 = np.linspace(-1, 1, 500)
y1 = porta(t1, .5)

t2 = np.linspace(-10, 10, 500)
y2 = np.sinc(t2)

y = y1 * y2
plt.plot(t2, y)
```
